```python
"""
TUGAS 
Nilai Akhir Ujian = 30% dari nilai UTS +  40% dari nilai UAS + 30 % dari TUGAS
Ketentuan :
a. Jika Nilai Akhir 60-70 -- > Kategori Rawan Mengulang & Bahaya
b. Jika Nilai Akhir 71-80 -- > Kategori Bagus
c. Jika Nilai Akhir 81-90 -- > Kategori Bagus Sekali
d. Jika Nilai Akhir 91-100 -- > Istimewa
e. Selain Nilai Diatas --> Kategori Harus Mengulang
"""


NilaiUTS   = float(input("Masukkan Nilai UTS   : "))
NilaiUAS   = float(input("Masukkan Nilai UAS   : "))
NilaiTugas = float(input("Masukkan Nilai Tugas : "))

NilaiAkhir = (NilaiUTS * 0.3) + (NilaiUAS * 0.4) + (NilaiTugas * 0.3)

print("=============================================")
print("Hasil Nilai Akhir    : ", NilaiAkhir)

if (60 < NilaiAkhir < 70):    
    print("Keterangan           :  Rawan Mangulang & Bahaya")

elif (71 < NilaiAkhir < 80):
    print("Keterangan           :  Bagus")

elif (81 < NilaiAkhir < 90):
    print("Keterangan           :  Bagus Sekali")
    
elif (91 < NilaiAkhir < 100):
    print("Keterangan           :  Istimewa")

else:
    print("Keterangan           :  Harus Mengulang")
```
